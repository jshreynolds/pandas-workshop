```python
import pandas as pd
import sqlite3

make_conn = lambda: sqlite3.connect('./datasets/supply_company.db')
curs = make_conn().cursor()

curs.execute("select name from sqlite_master where type='table';")

print(curs.fetchall())
```

```python
customers = pd.read_sql('select * from Customers', make_conn())
orders = pd.read_sql('select * from Orders', make_conn())
```

```python



total_sales = orders.groupby('Customer_Number').sum(numeric_only=True)['amount'].sort_values(ascending=False)

foo = pd.merge(customers, orders, on="Customer_Number" )

foo.rename(columns={"amount": "total_sales"})

```
