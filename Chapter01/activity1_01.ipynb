```python
import pandas as pd

store1 = pd.read_csv('./Datasets/Store1.csv')
store2 = pd.read_csv('./Datasets/Store2.csv')

store1.dtypes
```

```python
store1_sales_by_kind = store1[['Grocery_sales', 'Stationary_sales']].sum()
store1_total_sales = store1_sales_by_kind.sum()

store2_sales_by_kind = store2[['Grocery_sales', 'Stationary_sales']].sum()
store2_total_sales = store2_sales_by_kind.sum()

print(store1_total_sales > store2_total_sales)
#Store 1 had greater sales
```

```python
store1_sales_by_kind['Grocery_sales'] > store2_sales_by_kind['Grocery_sales']
#Store 2 had greater Grocery Sales
```
